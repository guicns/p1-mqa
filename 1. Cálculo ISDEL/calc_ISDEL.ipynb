Como calcular o valor da dimensão para cada município? Multiplicar, para cada município, o valor de uma subdimensão pelo peso atribuído à ela. Depois, somar as subdimensões que participam de uma dimensão. <br>

```python
pesos_sub = {
    "Densidade Empresarial": 0.2,
    "Renda": 0.3,
    "Educação": 0.5,
    "Valores Solidários - programas e ações": 0.3,
    "Tecido Social": 0.3,
    "Tecido Empresarial": 0.4,
    "Articulação": 0.2,
    "Gestão Pública": 0.4,
    "Participação e Controle Social": 0.4,
    "Vantagens Locais": 0.5,
    "Organização Produtiva": 0.5,
    "Comércio Internacional": 1,
}

pesos_sub_list = list(pesos_sub.values())
pesos_sub_list
#pesos_sub_list[0:3]
```

```python
df2 = df.copy()
```

```python
# CAPITAL EMPREENDEDOR
capital_empreendedor = df2.iloc[:,1]
df2.loc[:,"CAPITAL EMPREENDEDOR"] = capital_empreendedor.multiply(pesos_sub_list[0])

for i in range(1,3):
  capital_empreendedor = df2.iloc[:,1+i]
  df2.loc[:,"CAPITAL EMPREENDEDOR"] = df2.loc[:,"CAPITAL EMPREENDEDOR"] + capital_empreendedor.multiply(pesos_sub_list[i])
```

```python
# TECIDO EMPRESARIAL
tecido_empresarial = df2.iloc[:,4]
df2.loc[:,"TECIDO EMPRESARIAL"] = tecido_empresarial.multiply(pesos_sub_list[3])

for i in range(4,6): #4 to 5
  tecido_empresarial = df2.iloc[:,1+i]
  df2.loc[:,"TECIDO EMPRESARIAL"] = df2.loc[:,"TECIDO EMPRESARIAL"] + tecido_empresarial.multiply(pesos_sub_list[i])
```

```python
# GOVERNANÇA
governanca = df2.iloc[:,7]
df2.loc[:,"GOVERNANÇA"] = governanca.multiply(pesos_sub_list[6])

for i in range(7,9): # 7 to 8 (8 e 9)
  governanca = df2.iloc[:,1+i]
  df2.loc[:,"GOVERNANÇA"] = df2.loc[:,"GOVERNANÇA"] + governanca.multiply(pesos_sub_list[i])
```

```python
# ORGANIZAÇÃO PRODUTIVA
organizacao_produtiva = df2.iloc[:,10]
df2.loc[:,"ORGANIZAÇÃO PRODUTIVA"] = organizacao_produtiva.multiply(pesos_sub_list[9])

for i in range(10,11): # 10 to 11 (11)
  organizacao_produtiva = df2.iloc[:,1+i]
  df2.loc[:,"ORGANIZAÇÃO PRODUTIVA"] = df2.loc[:,"ORGANIZAÇÃO PRODUTIVA"] + organizacao_produtiva.multiply(pesos_sub_list[i])
```

```python
# INSERÇÃO COMPETITIVA
insercao_competitiva = df2.iloc[:,12]
df2.loc[:,"INSERÇÃO COMPETITIVA"] = insercao_competitiva.multiply(pesos_sub_list[11])
```

```python
pesos_isdel = {
    "CAPITAL EMPREENDEDOR": 0.25, #coluna 13
    "TECIDO EMPRESARIAL": 0.10, #coluna 14
    "GOVERNANÇA": 0.20, #coluna 15
    "ORGANIZAÇÃO PRODUTIVA": 0.25, #coluna 16
    "INSERÇÃO COMPETITIVA": 0.20, #coluna 17
}

pesos_isdel_list = list(pesos_isdel.values())
pesos_isdel_list
#pesos_sub_list[0:3]
```

```python
# Cálculo do ISDEL
ISDEL = df2.iloc[:,13]
df2.loc[:,"ISDEL"] = ISDEL.multiply(pesos_isdel_list[0])

for i in range(1,5): # 1 to 4 (coluna 14 até 17)
  ISDEL = df2.iloc[:,13+i]
  df2.loc[:,"ISDEL"] = df2.loc[:,"ISDEL"] + ISDEL.multiply(pesos_isdel_list[i])
```
